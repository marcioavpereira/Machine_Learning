```
import tensorflow as tf
import numpy as np
from IPython.display import Image as IImage, display
import PIL
from PIL import Image

tf.__version__
```

##Carregamento da Imagem

```
onca = PIL.Image.open('/content/Onça-pintada descansando.jpg')
onca.load()
onca_array = np.array(onca)
```

```
onca_array.shape
```

```
PIL.Image.fromarray(onca_array)
```

##Operações com a Imagem

###Rotação

```
onca_rotacionada = tf.image.random_flip_left_right(onca_array)
```

```
PIL.Image.fromarray(onca_rotacionada.numpy())
```

###Brilho

```
onca_brilho = tf.image.adjust_brightness(onca_array, delta=0.4)
PIL.Image.fromarray(onca_brilho.numpy())
```

###Redimensionamento

```
onca_resize = tf.image.resize(onca_array, size=(200,300), preserve_aspect_ratio=False)
onca_resize = tf.cast(onca_resize, tf.uint8)
PIL.Image.fromarray(onca_resize.numpy())
```
